# For Loops
### For loop is a block of code that repeats or iterates a task until certain conditions are met. They are present in all programming languages.


## Loops in Python

```python
for i in range(1,10):
  print(i)
```

In the code above, notice the key words "for" and "in". "i" is just a local variable, in fact "i" can be replaced by any other variable.

## Loops are very useful when dealing with Lists

```python
list_a = [1,2,3,4,5,6,7]
list_b = []
for val in list_a:
  list_b.append(val*2)

print(list_b)
```
